```r
library(prettyR)
library(Maaslin2)
```

```r
metadata = "../tables/subjects_metadata.tsv"
df_input_metadata = read.table(file = metadata, header = TRUE, sep = "\t", stringsAsFactors = FALSE, row.names = 2)

abundance = "../tmp/maaslin2/bracken_self_combine_result.genus_sum.matrix.tsv"
df_input_data = read.table(file = abundance, header = TRUE, sep = "\t", stringsAsFactors = FALSE, row.names = 1)
```

```r
fit_data2 = Maaslin2(
  input_data = df_input_data, 
  input_metadata = df_input_metadata, 
  normalization = "NONE",
  transform = "NONE",
  analysis_method = "CPLM",
  output = "../tmp/maaslin2/abundance_genus_mean_cplm_chinese", 
  fixed_effects = c("Age", "Sexe", "Ethnicity"),
  reference = c("Ethnicity,Chinese"))
```
