```r
## import packages
library(iNEXT)
library(ggplot2)
```

```r
data = read.table("../tmp/iNext/bracken_self_combine_result.matrix.tsv", sep="\t")
data = t(as.matrix(data))
data = apply(data, 1, as.logical)

data.list = list(data)
names(data.list) <- c("SPMP")

out.raw = iNEXT(data.list, q=0, datatype="incidence_raw", endpoint=300)
out.raw

ggiNEXT(out.raw) + geom_line(aes(y=out.raw$AsyEst[1, 4]), size = 0.5, linetype = "dashed", color = "black")
out.raw$AsyEst
```
