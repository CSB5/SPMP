```python
from ete3 import Tree

fname = "../tables/gtdbtk.bac120.classify.tree"
tree = Tree(fname, format=1, quoted_node_names=True)

keep = []

for node in tree.traverse("postorder") :
    name = node.name
    if node.is_leaf() : continue
    
    if "o__" not in name : continue
    
    if not any(leaf.name.startswith("TLL")
          for leaf in node.get_leaves()) : continue
    
    keep.append(node)
    for child in node.children :
        node.remove_child(child)
    
    name = name.split(":")[1]
    name = dict(sub.split("__") for sub in name.split(";"))
    print (name["o"])
    
tree.prune(keep)
print (tree)
# arc : Methanobacteriales
```
